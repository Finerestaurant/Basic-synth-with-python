```python
import numpy as np
from scipy import signal
import pandas as pd 
from IPython.display import Audio
import matplotlib.pyplot as plt
import librosa.display
```

```python
data = pd.read_csv("/content/pitch.csv")

Note = data.drop([data.columns[2]], axis=1)
pitch_info_pre = Note.to_dict() 

keys = list(pitch_info_pre['Note'].values())
frequency_info = list(pitch_info_pre['Frequency (Hz)'].values())

for i in range(len(keys)-1):
    keys[i] = keys[i].replace(u'\xa0', u' ')

pitch_info = {}
for i in range(len(keys)-1):
    pitch_info[keys[i]] = frequency_info[i]
```

```python
def create_scale(key, modes="major"):
    interval_bench = keys.index(key)
    scale_num_list = []
    scale_keys_list = []
    one_octave = 12
    for i in range(0,one_octave*2):       
        if modes == "major":
           if i % 12 == 1 or i % 12 == 3 or i % 12 == 6 or i % 12 == 8 or i % 12 == 10:
              continue
        elif modes == "minor":
           if i % 12 == 1 or i % 12 == 4 or i % 12 == 5 or i % 12 == 9 or i % 12 == 11:
              continue     
        scale_num_list.append(interval_bench+i)
        
    for i in scale_num_list:
        scale_keys_list.append(keys[i])
    return scale_keys_list

def square_wave(frequency):
    wave = signal.square(2*np.pi*frequency*t)
    return wave

def sawtooth_wave(frequency):
    wave = signal.sawtooth(2*np.pi*frequency*t)
    return wave

def sine_wave(frequency):
    wave = np.sin(2*np.pi*frequency*t)
    return wave
```

```python
framerate = 44100
time_range = 15
t = np.linspace(0,time_range,framerate*time_range)
amplitude = 1
scale =  input("'major' or 'minor'  ")
key =  input("Keys: ex: D3   ")
chord = list(map(int, input("interval : put numbers here  ").split(" ")))
wave = input("Wave : sine, square, sawtooth")
print(chord)
scale = create_scale(key, scale)
print('major',scale)

def chord_list(input_list):
    chord_list = []
    for i in input_list:
        chord_list.append(scale[i-1])
    return chord_list  

chord = chord_list(chord)

freq = []
for i in chord:
   freq.append(pitch_info[i])
   print(i)
print(freq)

total = np.zeros(framerate*time_range,) 
for i in freq:
    if wave == "sine":
       total += sine_wave(i)
    elif wave == "square":
       total += square_wave(i)
    elif wave == "sawtooth":
       total += sawtooth_wave(i)


Audio(total,rate = framerate)


```
